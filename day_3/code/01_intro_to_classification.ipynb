Classification is generally trained through supervised, though semi-supervised learning can also be used.  While there are relatively few algorithms for regression there are many different types of classification algorithm.

#### Binary vs. Multi-class

One way to stratify classification problems is the size of the set of the target class. This is either binary (0 or 1) or multi-class (Group A, Group B, 'Group C)

Most algorithms support both, however in some instances (such as support vector machines) are combinations of binary classifiers.

### Types of Classification

#### Eager vs. Lazy

**Lazy Learning** stores training data without further processing. It awaits a test set to then classify. The most popular example of this is K-Nearest Neighbors (KNN).

**Eager Learning** takes the provided training data and constructs an algorithm specific to those training data.  When new observations from a test set arrive, they are classified using that constructed algorithm. Examples include random forests and logistic regression.

### The Confusion Matrix

A confusion matrix is a diagnostic tool to understand classification model performance. Many of the common measures we use to evaluate can be derived from this matrix.  The rows denote the true class, and the columns the predicted.

```python
from sklearn.metrics import confusion_matrix
y_true = ["beaver", "beaver", "sloth", "sloth", "sloth"]
y_pred = ["beaver", "sloth", "sloth", "sloth", "sloth"]
print(confusion_matrix(y_true, y_pred, labels=["beaver", "sloth"]))
```

If we consider the sloth to be the target class, what matrix element represents the false positives?

```python
y_true = ["beaver", "beaver", "sloth", "sloth", "sloth","dog","dog"]
y_pred = ["beaver", "sloth", "sloth", "sloth", "dog","dog", "sloth"]
print(confusion_matrix(y_true, y_pred, labels=["beaver", "sloth","dog"]))
```
